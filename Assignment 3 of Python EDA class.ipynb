Q8. Using NumPy, create a function to normalize the 'BloodPressure' values (subtract the
mean and divide by the standard deviation).

```python
def normalize_blood_pressure(blood_pressure):
    mean_pressure = np.mean(blood_pressure)
    std_pressure = np.std(blood_pressure)
    normalized_pressure = (blood_pressure - mean_pressure) / std_pressure
    return normalized_pressure

normalized_blood_pressure = normalize_blood_pressure(data['BloodPressure'].values)
print(f"Normalized Blood Pressure values:\n{normalized_blood_pressure}")
```

Q4. Create a new column 'BMI_Class' categorizing BMI values into 'Underweight', 'Normal',
'Overweight', and 'Obese'.

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

data['BMI_Class'] = data['BMI'].apply(categorize_bmi)
```
